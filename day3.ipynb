```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
import numpy as np

(X_train -np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()

X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car',as_frame=True)
```

```python
data
```

```python
data = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
columns_to_encode = ['lug_boot','safety']

encoder = OrdinalEncoder(
    categories=[
        ['small','med','big'],
        ['low','med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
data
```

```python
encoder.inverse_transform(data[columns_to_encode])
```
