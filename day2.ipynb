```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()

X,y = data.data,data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```
